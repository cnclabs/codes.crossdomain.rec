```python
# tv
```

```python
tv_interaction_log = pd.read_csv('../../CPR/input/all_tv_train_input.txt', header=None, sep='\t')
```

```python
tv_interaction_log['user_embedding'] = 0
tv_interaction_log['item_embedding'] = 0
tv_interaction_log.columns=['userID', 'itemID', 'rating', 'user_embedding', 'item_embedding']
tv_interaction_log.head()
```

```python
tv_emb = pd.read_csv('../BPR/graph/tv_lightfm_bpr_10e-5.txt', header=None, sep=' ', skiprows=1)
tv_emb.head()
```

```python
tv_emb_dict = {}
for i in range(len(tv_emb)):
    k = tv_emb.iloc[i, 0]
    v = list(tv_emb.iloc[i, 3:].values)
    tv_emb_dict[k] = v
```

```python
tv_interaction_log['user_embedding'] = tv_interaction_log['userID'].map(tv_emb_dict)
tv_interaction_log['item_embedding'] = tv_interaction_log['itemID'].map(tv_emb_dict)
```

```python
tv_interaction_log.head()
```

```python
with open('./input/tv_lightfm.pickle', 'wb') as pf:
    pickle.dump(tv_interaction_log, pf)
```
